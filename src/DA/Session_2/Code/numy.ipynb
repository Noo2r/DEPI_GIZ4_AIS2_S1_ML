```python
"""Reshape a numpy array to the specified number of rows and columns.
    
    Parameters:
        arr (np.ndarray): The input array to reshape.
        rows (int): The desired number of rows. Default is -1 (unspecified).
        cols (int): The desired number of columns. Default is -1 (unspecified).
        
    Returns:
        np.ndarray: The reshaped array."""
    
def array_reshape(arr, rows = -1, cols = -1):
    try:
        arr = np.array(arr)
        reshaped_arr = arr.reshape(rows, cols)
        return reshaped_arr
    except ValueError:
        print("Error: Invalid shape. The total number of elements in the array must remain the same.")
        return None
```

```python
vec = np.array([1, 2, 3, 4, 5, 6])
print("Original Vector:", vec)
reshaped_matrix = array_reshape(vec, rows=2, cols=3)
print("Reshaped Matrix:")
print(reshaped_matrix)
```

```python
v = np.array([10, 25, 37, 54, 65]) 
condition = v >= 40
print(condition)
```

```python
v = np.array([10, 25, 37, 54, 65]) 
condition = v >= 40
conditional_access = v[condition]
print(condition)
print("Values meeting the condition:", conditional_access)
```

```python
v = np.array([10, 25, 37, 54, 65]) 
condition = v >= 40
new_array = np.where(condition, v, -1)
print("Original array:", v)
print("Condition array:", condition)
print("New array with -1 where condition is False:", new_array)
```
