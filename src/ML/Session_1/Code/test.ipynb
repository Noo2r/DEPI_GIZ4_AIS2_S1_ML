```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2,2.2,3.6,4.7])
```

```python
w = 0 # weight
b = 0 # bias
alpha = 0.01 # learning rate
num_iters = 20
```

```python
sse_values = []
for i in range(num_iters):
    y_pred = w * x + b
    error = y_pred - y
    sse = np.sum(error ** 2)
    sse_values.append(sse)

    dw = (2 / len(x)) * np.dot(error, x) # sum(error * x)
    db = (2 / len(x)) * np.sum(error)

    w -= alpha * dw
    b -= alpha * db

    if (i+1) % 20 == 0:
        print(f"Iteration {i+1}: w={w:.4f}, b={b:.4f}, SSE={sse:.4f}")  
```

```python
print(sse_values)
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(range(num_iters),sse_values, label='SSE' ,color='green')
plt.xlabel('Iteration')
plt.ylabel('SSE')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x, y, label='Data Points', color='red')
plt.plot(x, w * x + b, label='Fitted Line', color='blue')

plt.show()
```

Class design named Linear_Regression

```python
class Linear_Regression_Nour_Eldin_Ahmed:
    def __init__(self, learning_rate=0.01, num_iters=1000):
        self.learning_rate = learning_rate
        self.num_iters = num_iters
        self.w = 0  
        self.b = 0  

    def fit(self, X, y):
        n_samples = len(X)
        sse_values = []

        for _ in range(self.num_iters):
            y_predicted = self.w * X + self.b
            sse = np.sum((y - y_predicted) ** 2)
            sse_values.append(sse)

            dw = (-2 / n_samples) * np.sum(X * (y - y_predicted))
            db = (-2 / n_samples) * np.sum(y - y_predicted)

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

        return sse_values
```
