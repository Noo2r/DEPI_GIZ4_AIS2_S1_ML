## Task-1:
Suppose you have this email address “Amit_ml@gmail.edu”

```python
email = input("Enter your email address: ")
count = email.count("@")
if count == 1 and "." in email.split("@")[-1]:
    print("Valid email address.")  
    print("Username:", email.split("@")[0])
    print("Domain:", email.split("@")[-1].split(".")[0])
    if email.endswith(".com"):
        print("Commercial Domain.")
    elif email.endswith(".oedu"):
        print("Educational Domain.")
    else:
        print("Other Domain.")     
else:
    print("Invalid email address.")

    
```

## Task-2: 
Encoded Message: 
###!!@mocleW EPGTQ!!!6789 

```python
Encoded_Message = "###!!@mocleW EPGTQ!!!6789" 
Extract_message = Encoded_Message[6:18]
word_1 = Extract_message[0:6]
reversed_word_1 = word_1[::-1]
word_2 = Extract_message[8:12]
decoded_message = reversed_word_1 + " " + word_2
print("Decoded Message:", decoded_message)
```

## Task-3: 
Encoded Message: 
&&&**$gnirtS PLIO!!@1234

```python
Encoded_Message = "&&&**$gnirtS PLIO!!@1234" 
Extract_message = Encoded_Message[6:17]
word_1 = Extract_message[0:6]
reversed_word_1 = word_1[::-1]
word_2 = Extract_message[7:12]
replaced_word_2 = word_2.replace("I", "E").replace("O", "U")
decoded_message = reversed_word_1 + " " + replaced_word_2
print("Decoded Message:", decoded_message)
```

## Task-4: 
Encoded Message: 
##$$$@!yalpstcejorp EPUVT****9887

```python
Encoded_Message = "##$$$@!yalpstcejorp EPUVT****9887" 
Extract_message = Encoded_Message[7:25]
word_1 = Extract_message[0:12]
reversed_word_1 = word_1[::-1]
word_2 = Extract_message[13:19]
replaced_word_2 = word_2.replace("E", "A").replace("U", "O")
decoded_message = reversed_word_1 + " " + replaced_word_2
print("Decoded Message:", decoded_message)
```
